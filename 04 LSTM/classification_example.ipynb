```python
import pathlib, sys
import numpy as np

from lstm import LSTMClassifier
```

```python
rng = np.random.default_rng(42)
N = 240
T = 8
D = 2

# простой датасет: метка = 1, если сумма значений в последовательности > 0
X = rng.normal(size=(N, T, D))
y = (np.sum(X, axis=(1, 2)) > 0).astype(int)

split = int(0.8 * N)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
model = LSTMClassifier(
    input_dim=D,
    hidden_dim=32,
    output_dim=2,
    lr=1e-2,
    max_epochs=50,
    random_state=0,
    verbose=False,
)

model.fit(X_train, y_train)
print("train acc:", model.score(X_train, y_train))
print("test  acc:", model.score(X_test, y_test))
```

```python
# прогноз на одном примере
sample_idx = 0
probs = model.predict_proba(X_test[sample_idx:sample_idx+1])[0]
print("probs", probs)
print("pred label", np.argmax(probs), "true", y_test[sample_idx])
```
